```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import os


# Set your working directory
data_path = "c:/Users/CESAR/Desktop/Encuentro economistas 2024 - Propuesta/datos/EH_DTA/"
os.chdir(data_path)
```

### 2023

```python
# Load the dataset
df = pd.read_stata('EH2023_Persona.dta')

# Inequality Indicators

# Cleaning data
df = df[(df['yhogpc'] != 0) & (~df['yhogpc'].isnull())]

# Calculating deciles and percentiles
perc10_v = np.percentile(df['yhogpc'], 10)
perc100_v = np.percentile(df['yhogpc'], 100)

# Creating deciles and percentiles
df['perc100'] = pd.qcut(df['yhogpc'], 100, labels=False)
df['perc10'] = pd.qcut(df['yhogpc'], 10, labels=False)

# Estimating income distribution
income_dist_perc10 = df.groupby('perc10')['yhogpc'].agg(['mean', 'median', 'min', 'max', 'count', 'sum'])
income_dist_perc100 = df.groupby('perc100')['yhogpc'].agg(['mean', 'median', 'min', 'max', 'count', 'sum'])

# Saving the results to Excel
# We will create a excel, then we create a sheet for each Household Survey Year
# Income Distribution (Perc10) / (Perc10) id_10 
with pd.ExcelWriter('Income_dist_EH_History.xlsx') as writer:
    income_dist_perc10.to_excel(writer, sheet_name='yd010_2023')
    income_dist_perc100.to_excel(writer, sheet_name='yd100_2023')
    
```

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar el archivo Excel con los datos de ingreso por hogar
file_path_lorenz = "df_combined.xlsx"  # Asegúrate de tener el archivo en tu directorio
xls_lorenz = pd.ExcelFile(file_path_lorenz)

# Cargar la hoja de datos
df_lorenz = pd.read_excel(xls_lorenz, sheet_name=xls_lorenz.sheet_names[0])

# Eliminar la primera columna innecesaria (si es un índice)
df_lorenz = df_lorenz.iloc[:, 1:]

# Función para calcular la curva de Lorenz
def calculate_lorenz_curve(df):
    df_sorted = df.sort_values('yhogpc')
    n = len(df_sorted)
    cum_income = np.cumsum(df_sorted['yhogpc'])
    cum_percent = 100 * np.arange(1, n + 1) / n
    lorenz_curve = pd.DataFrame({'cum_percent': cum_percent, 'cum_income': cum_income})
    lorenz_curve['norm_cum_income'] = lorenz_curve['cum_income'] / lorenz_curve['cum_income'].max()
    lorenz_curve['norm_cum_percent'] = lorenz_curve['cum_percent'] / lorenz_curve['cum_percent'].max()
    return lorenz_curve

# Función para calcular el coeficiente de Gini
def calculate_gini(df):
    if df['yhogpc'].nunique() <= 1:  # Si no hay variación en los ingresos, devolver NaN
        return None
    sorted_values = np.sort(df['yhogpc'])
    n = len(sorted_values)
    cumulative_income = np.cumsum(sorted_values)
    cumulative_income = np.insert(cumulative_income, 0, 0)

    # Calcular el área bajo la curva de Lorenz usando la regla del trapecio
    B = np.trapz(cumulative_income / cumulative_income[-1], dx=1/n)
    
    # El coeficiente de Gini se calcula como A/(A+B), donde A+B = 0.5
    gini_index = 1 - 2 * B
    return gini_index

# Calcular el coeficiente de Gini para cada año
gini_values = {year: calculate_gini(data) for year, data in df_lorenz.groupby('year')}

# Configurar Matplotlib para usar una fuente académica
```
