```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# Data for the plot
countries = ["Chile", "Ecuador", "Bolivia", "Peru", "Uruguay", "Paraguay", "Brasil", "Argentina", "Colombia"]
changes_2022 = [1.70, 1.47, 1.04, 1.02, 0.95, 0.80, 0.76, 0.59, 0.48]
changes_2021 = [2.97, 0.77, 0.48, 2.80, -1.02, 0.36, 2.05, -0.70, 0.43]

# Normalize values for a lighter grayscale gradient effect
normalized_values_2022 = (np.array(changes_2022) - min(changes_2022)) / (max(changes_2022) - min(changes_2022))
lighter_colors_2022 = [(0.5 + 0.4 * norm, 0.5 + 0.4 * norm, 0.5 + 0.4 * norm) for norm in normalized_values_2022]

normalized_values_2021 = (np.array(changes_2021) - min(changes_2021)) / (max(changes_2021) - min(changes_2021))
lighter_colors_2021 = [(0.5 + 0.4 * norm, 0.5 + 0.4 * norm, 0.5 + 0.4 * norm) for norm in normalized_values_2021]

# Create figure and axis with high resolution
fig, ax = plt.subplots(figsize=(8, 5), dpi=300)

# Plot horizontal bar chart with lighter grayscale gradient for 2022 changes
bars = ax.bar(countries, changes_2022, color=lighter_colors_2022, edgecolor="black")

# Apply texture (hatching) to Bolivia
for bar, country in zip(bars, countries):
    if country == "Bolivia":
        bar.set_hatch("//////")  # Denser diagonal hatching

# Add larger circles for 2021 values with lighter grayscale colors
circle_radius = 0.35  # Keep circles large for readability
for i, country in enumerate(countries):
    circle = plt.Circle((i, changes_2021[i]), circle_radius, color=lighter_colors_2021[i], ec="black", lw=1.5, zorder=3)
    ax.add_patch(circle)
    # Add text inside the circle
    ax.text(i, changes_2021[i], f"{changes_2021[i]:.2f}", fontsize=10,
            ha='center', va='center', color="black")

# Add text labels for bars (2022), ensuring proper spacing
for bar, value in zip(bars, changes_2022):
    ax.text(bar.get_x() + bar.get_width()/2, value + 0.4, f"{value:.2f}", 
            ha='center', fontsize=10, color="black")

# Adjust x-axis labels
ax.set_xticks(range(len(countries)))
ax.set_xticklabels(countries, rotation=45, ha="right")

# Set y-axis limit to 4 as requested
ax.set_ylim(-2.5, 4)

# Labels
ax.set_ylabel("Percentage Points", fontsize=12)

# Improve aesthetics
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Save the figure in high resolution
output_path = "tax_changes_2021_2022_ymax4.png"
plt.savefig(output_path, dpi=600, bbox_inches='tight', transparent=True)

# Display the final improved plot
plt.show()
```
